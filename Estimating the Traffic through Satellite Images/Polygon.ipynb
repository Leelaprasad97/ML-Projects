```python
import shapefile
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

# Load the Shapefile for Houston zip codes
sf = shapefile.Reader("tl_2020_us_zcta510/tl_2020_us_zcta510.shp")

# Get the index of the "ZCTA5CE10" field
zip_field_idx = [f[0] for f in sf.fields].index("ZCTA5CE10") - 1

# Define the coordinates of the bounding box for Houston
lon_min, lat_min = -95.7998, 29.5235
lon_max, lat_max = -95.2557, 30.1102

# Define the list of zip codes to plot
zip_codes = ['77002', '77003', '77004', '77005', '77006', '77007', '77008', '77009', '77010', '77011', '77012', '77013', '77014', '77015', '77016', '77017', '77018', '77019', '77020', '77021', '77022', '77023', '77024', '77025', '77026', '77027', '77028', '77029', '77030', '77031', '77032', '77033', '77034', '77035', '77036', '77037', '77038', '77039', '77040', '77041', '77042', '77043', '77044', '77045', '77046', '77047', '77048', '77049', '77050', '77051', '77053', '77054', '77055', '77056', '77057', '77058', '77059', '77060', '77061', '77062', '77063', '77064', '77065', '77066', '77067', '77068', '77069', '77070', '77071', '77072', '77073', '77074', '77075', '77076', '77077', '77078', '77079', '77080', '77081', '77082', '77083', '77084', '77085', '77086', '77087', '77088', '77089', '77090', '77091', '77092', '77093', '77094', '77095', '77096', '77098', '77099']

# Loop through the zip codes in the list and extract their polygon coordinates
for zip_code in zip_codes:
    # Get the shape record for the zip code
    shape_records = sf.shapeRecords()
    shape_record = next(sr for sr in shape_records if sr.record[zip_field_idx] == zip_code)
    
    # Get the polygon coordinates for the zip code
    polygon = shape_record.shape.points
    
    # Create a Polygon patch for the zip code
    poly_patch = Polygon(polygon, facecolor='none', edgecolor='blue')

    # Create a new plot for the zip code
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.add_patch(poly_patch)

    # Set the axis limits and title
    ax.set_xlim([lon_min, lon_max])
    ax.set_ylim([lat_min, lat_max])
    ax.set_title("Zip Code: {}".format(zip_code))

    # Print the polygon coordinates
    print("Zip Code: {}".format(zip_code))
    for p in polygon:
        print("  ({:.6f}, {:.6f})".format(p[0], p[1]))

    # Show the plot
    plt.show()
```
